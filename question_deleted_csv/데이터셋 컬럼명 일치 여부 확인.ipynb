```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import glob
```

```python
# !!! 아래 코드를 실행하기 전 확인사항 !!!
# 현재 jupyternotebook 실행파일이 csv 데이터셋들과 같은 디렉토리 안에 있어야 함
file_list = glob.glob('*.csv')
file_list
```

```python
# 25개의 csv 파일을 DF로 변환하여 저장할 변수 25개 list로 묶기
df_list = list(range(25))
df_list
```

```python
# csv 파일 하나씩 읽어와 df_list에 차례로 대입
for i, file_name in enumerate(file_list):
    df_list[i] = pd.read_csv(file_name, encoding='cp949')
```

```python
# 결과 확인
df_list
```

```python
# df_list에 저장된 각각의 DF에 column명만 불러오기 예시
df_list[0].columns
```

```python
# 각각의 DF의 컬럼명이 같은지 비교
for i, df in enumerate(df_list):
    if i < 24:
        # i번쨰 DF 컬럼과 i+1번째 DF 컬럼을 비교한 불리언 배열에 False가 있는지 출력
        print(False in df_list[i].columns == df_list[i+1].columns ) 
    else:
        break
```
